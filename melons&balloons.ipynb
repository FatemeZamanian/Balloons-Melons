```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=1000
balloons_weight=np.random.normal(60,5,N)
balloons_width=np.random.normal(15,5,N)
balloons_height=np.random.normal(25,5,N)

melon_weight=np.random.normal(500,50,N)
melon_width=np.random.normal(15,5,N)
melon_height=np.random.normal(30,10,N)

width=np.concatenate((balloons_width,melon_width))
height=np.concatenate((balloons_height,melon_height))
weight=np.concatenate((balloons_weight,melon_weight))

x=np.array([width,height,weight]).T

melon_label=np.zeros(N,dtype='int')
balloons_label=np.ones(N,dtype='int')
Y=np.concatenate((melon_label,balloons_label))
```

```python
from mpl_toolkits.mplot3d import Axes3D
p=plt.figure()
fig=Axes3D(p)
fig.scatter(x[0:N,0],x[0:N,1],x[0:N,2], c='blue')
fig.scatter(x[N:,0],x[N:,1],x[N:,2], c='pink')

fig.set_xlabel('widths')
fig.set_ylabel('lengths')
fig.set_zlabel('weights')
plt.show()
```
